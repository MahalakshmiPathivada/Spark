1. Question
List all customers who have placed at least one order.
- Explanation:
_We need only those customers who have matching orders._

Expected result

- 1 | Krishna | Hyderabad
- 2 | Sudheer | Bangalore
- 4 | Rahul | Hyderabad

```sql

SELECT c.customer_id, c.customer_name, c.city
FROM Customer c
INNER JOIN Orders o
ON c.customer_id = o.customer_id;
```

## 2. Question
List all customers who have never placed any order.
> Explanation:
_We need all customers, including those without any orders._

```sql
SELECT c.customer_id, c.customer_name, c.city
FROM Customer c
LEFT JOIN Orders o
ON c.customer_id = o.customer_id
WHERE o.order_id IS NULL;
```

## 3. Question
List customers and their order count (including customers with no orders).
> Explanation:

We need all customers and count how many orders each has.


Expected Result
- 1 | Krishna | Hyderabad | 2
- 2 | Sudheer | Bangalore | 2
- 3 | Atul | Pune | 0
- 4 | Rahul | Hyderabad | 2

```sql
SELECT c.customer_id, c.customer_name, c.city,
       COUNT(o.order_id) AS order_count
FROM Customer c
LEFT JOIN Orders o
ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.customer_name, c.city
ORDER BY c.customer_id;
```

## 4. Question
Find total amount spent by each customer.
> Explanation:
We need total order amount for each customer.

Expected Result
- 1 | Krishna | 4000
- 2 | Sudheer | 4500
- 3 | Atul | 0
- 4 | Rahul | 3400

```sql
SELECT c.customer_id, c.customer_name,
       COALESCE(SUM(o.amount), 0) AS total_amount
FROM Customer c
LEFT JOIN Orders o
ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.customer_name
ORDER BY c.customer_id;
```

## 5. Question
List all employees who belong to the 'HR' department.
> Explanation:
We need employees who are in the HR department.

```sql
SELECT e.name, d.department_name
FROM Employee e
INNER JOIN Department d
ON e.department_id = d.department_id
WHERE d.department_name = 'HR';
```

## 6. Question
List the number of employees in each department.
- Explanation:
We need to count employees grouped by each department.

Expected Output
- Finance | 1
- HR | 2
- IT | 1
- Sales | 1

```sql
SELECT d.department_name,
       COUNT(e.employee_id) AS employee_count
FROM Department d
LEFT JOIN Employee e
ON d.department_id = e.department_id
GROUP BY d.department_name
ORDER BY d.department_name;
```

# Banking

```sql
-- Customers Temp View
CREATE OR REPLACE TEMP VIEW banking_customers AS
SELECT * FROM VALUES
(101, 'Krishna', 'krishna@gmail.com'),
(102, 'Sudheer', 'sudheer@gmail.com'),
(103, 'Raj', 'raj@gmail.com'),
(104, 'Ashna', 'ashna@gmail.com'),
(105, 'Parvati', 'parvati@gmail.com')
AS customers(customer_id, name, email);


-- Accounts Temp View
CREATE OR REPLACE TEMP VIEW accounts AS
SELECT * FROM VALUES
(201, 101, 'Savings', 25000.00),
(202, 102, 'Current', 15000.00),
(203, 103, 'Savings', 32000.00),
(204, 104, 'Current', 18000.00),
(205, 105, 'Savings', 27500.00)
AS accounts(account_id, customer_id, account_type, balance);


-- Transactions Temp View
CREATE OR REPLACE TEMP VIEW transactions AS
SELECT * FROM VALUES
(301, 201, DATE '2024-05-01', 'Deposit', 5000.00),
(302, 202, DATE '2024-05-02', 'Withdrawal', 3000.00),
(303, 201, DATE '2024-05-03', 'Withdrawal', 2000.00),
(304, 203, DATE '2024-05-04', 'Deposit', 4000.00),
(305, 204, DATE '2024-05-05', 'Deposit', 6000.00),
(306, 205, DATE '2024-05-06', 'Withdrawal', 1500.00)
AS transactions(txn_id, account_id, txn_date, txn_type, amount);
```

## 13. Question

- Krishna | Savings | 25000.00

- Raj | Savings | 32000.00
Sudheer | Current | 15000.00

```sql
SELECT c.name, a.account_type, a.balance
FROM banking_customers c
JOIN Accounts a ON c.customer_id = a.customer_id
ORDER BY c.name;
```

## 14. Question
Find the total amount deposited by each customer.

Expected Output
- Raj | 4000.00
- Ashna | 6000.00
- Krishna | 5000.00
- Sudheer | 0.00
- Parvati | 0.00

```sql
SELECT c.name, COALESCE(SUM(t.amount), 0) AS total_deposited
FROM banking_customers c
JOIN Accounts a ON c.customer_id = a.customer_id
LEFT JOIN Transactions t
ON a.account_id = t.account_id AND t.txn_type = 'Deposit'
GROUP BY c.name
ORDER BY total_deposited DESC;
```

Find the average salary of employees in each department.

Expected Output

- Finance | 65000.00

## 16. Question
List the employees who received a rating of 4 or above in their latest performance review.

Expected Output
- Ashna | 5 | 2024-06-30
- Parvati | 5 | 2024-06-30
- Pinki | 4 | 2024-06-30
- Shruti | 4 | 2024-06-30

```sql
SELECT e.name, p.rating, p.review_date
FROM Employees e
JOIN Performance p ON e.emp_id = p.emp_id
WHERE (e.emp_id, p.review_date) IN (
    SELECT emp_id, MAX(review_date)
    FROM Performance
    GROUP BY emp_id
)
AND p.rating >= 4
ORDER BY e.name;
```
